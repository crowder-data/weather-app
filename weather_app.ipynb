```python
import requests

normal_params =  {
    "datasetid": "NORMAL_DLY",
    "stationid": "GHCND:USW00014920",
    "datatypeid": ["DLY-TMAX-NORMAL","DLY-TMIN-NORMAL"],
    "startdate": "2010-01-01",
    "enddate": "2010-12-31",
    "limit": 1000
}


actual_params = {
    "datasetid": "GHCND",
    "stationid": "GHCND:USW00014920",
    "datatypeid": ["TMAX", "TMIN"],
    "startdate": "2025-01-01",
    "enddate": "2025-12-31",
    "limit": 1000
}


url = "https://www.ncei.noaa.gov/cdo-web/api/v2/data"


headers = {
    "token": token
}


actuals = requests.get(url, headers=headers, params = actual_params)
data_actuals = actuals.json()
normals = requests.get(url, headers=headers, params = normal_params)
data_normals = normals.json()
```

```python
#normals
df = pd.DataFrame(data_normals["results"])
df["date"] = pd.to_datetime(df["date"])
df["Month"] = df["date"].dt.month_name()
df["Day"] = df["date"].dt.day
df.drop(columns=['station','attributes','date'],inplace=True)
df['datatype'] = df['datatype'].replace({'DLY-TMAX-NORMAL':'Normal High','DLY-TMIN-NORMAL':'Normal Low'})
df = df.pivot(index = ['Month','Day'],columns = 'datatype', values = 'value').reset_index()
df[["Normal High", "Normal Low"]] = df[["Normal High", "Normal Low"]] / 10
df
```

```python
df2 = pd.DataFrame(data_actuals["results"])
df2["Date"] = pd.to_datetime(df2["date"])
df2["Month"] = df2["Date"].dt.month_name()
df2["Day"] = df2["Date"].dt.day
df2["Date"] = df2["Date"].dt.date
df2.drop(columns=['station','attributes'],inplace=True)
df2 = df2.pivot(index = ['Date','Month','Day'],columns = 'datatype', values = 'value').reset_index()
df2[["TMAX", "TMIN"]] = ((df2[["TMAX", "TMIN"]] / 10) * 9/5 + 32).round(1)
df2.rename(columns = {'TMAX':'High','TMIN':'Low'},inplace=True)
df2
```

```python
df_combined = df2.merge(df,on=['Month','Day'],how='left',suffixes = ("_actual","_normal"))
```

```python
df_combined = df_combined.iloc[:,[0,1,2,3,5,4,6]]
df_combined
```

# Thresholds

```python
# df_combined['Act-Norm High'] = df_combined['High'] - df_combined['Normal High']
# df_combined['Act-Norm Low'] = df_combined['Low'] - df_combined['Normal Low']

bins = [0, 5, 10, 15, 20, float("inf")]

df_combined["High diff"] = pd.cut(
    df_combined["High"] - df_combined["Normal High"],
    bins=bins,
    right=False
)

df_combined["Low diff"] = pd.cut(
    df_combined["Low"] - df_combined["Normal Low"],
    bins=bins,
    right=False
)
```

```python
df_combined
```

```python
high_counts = df_combined["High diff"].value_counts()
low_counts = df_combined["Low diff"].value_counts()

counts = pd.DataFrame({
    "High": high_counts,
    "Low": low_counts
}).fillna(0)

plot_df = counts.T

ax = plot_df.plot(
    kind="bar",
    stacked=True,
    color=["mistyrose", "lightcoral", "indianred", "firebrick", "darkred"],
    title = "Days above Normal Temperatures by Threshold for La Crosse, WI (2025)",
    figsize=(8,5)
)

for container in ax.containers:
    ax.bar_label(container, label_type="center")

ax.set_xlabel("Type")
ax.set_ylabel("Number of Days")

ax.legend(
    title="Exceedance Range (F)",
    bbox_to_anchor=(1.05, 1),
    loc="upper left"
);
```

```python
deviations = [0,5,10,15,20]
high_counts = []
low_counts = []
for each in deviations:
  high_counts.append(((df_combined['High']-df_combined['Normal High'])>each).sum())
  low_counts.append(((df_combined['Low']-df_combined['Normal Low'])>each).sum())
deviation_df = pd.DataFrame({
    "actual-normal thresh (F)":deviations,
    "highs above thresh":high_counts,
    "lows above thresh":low_counts
})
```

```python
deviation_df
```

```python
deviation_df['highs'] = deviation_df['highs above thresh'].diff(periods=-1).fillna(deviation_df['highs above thresh'])
deviation_df['lows'] = deviation_df['lows above thresh'].diff(periods=-1).fillna(deviation_df['lows above thresh'])
deviation_df
```

```python
df_thresh = deviation_df.iloc[:,[0,3,4]]
df_thresh = df_thresh.set_index('actual-normal thresh (F)').T
plot_df = df_thresh
plot_df
```

```python
ax = plot_df.plot(
    kind="bar",
    stacked=True,
    color=["mistyrose", "lightcoral", "indianred", "firebrick", "darkred"],
    title = "Days above Normal Temperatures by Threshold for La Crosse, WI (2025)",
    figsize=(8,5)
)

for container in ax.containers:
    ax.bar_label(container, label_type="center")

ax.set_xlabel("Type")
ax.set_ylabel("Number of Days")

ax.legend(
    title="Exceedance Floor (F)",
    bbox_to_anchor=(1.05, 1),
    loc="upper left"
);
```

# App with "felt like" matches

```python
df_combined
```

```python
def find_high_match(row):
    diff = (df_combined["Normal High"] - row["High"]).abs()
    return df_combined.loc[diff.idxmin(), 'Month'] + ' ' + str(df_combined.loc[diff.idxmin(), 'Day'])

df_combined["High Felt Like"] = df_combined.apply(find_high_match, axis=1)
```

```python
def find_low_match(row):
    diff = (df_combined["Normal Low"] - row["Low"]).abs()
    return df_combined.loc[diff.idxmin(), 'Month'] + ' ' + str(df_combined.loc[diff.idxmin(), 'Day'])

df_combined["Low Felt Like"] = df_combined.apply(find_low_match, axis=1)
```

```python
df_combined_app = df_combined.iloc[:,[1,2,3,4,7,5,6,8]]
df_combined_app
```
